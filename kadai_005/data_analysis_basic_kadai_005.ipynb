```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(2, 3, figsize=(15, 10), tight_layout=True)

y0 = [1, 2, -5, 2]
axes[0, 0].plot(y0)
axes[0, 0].grid(True)

x1 = np.linspace(0, 10, 30)
y1 = 2 + 2 * np.sin(2 * x1)
axes[0, 1].plot(x1, y1, linewidth=2.0)
axes[0, 1].set_xlim(0, 10)
axes[0, 1].set_ylim(-1, 5)
axes[0, 1].grid(True)

x2 = np.random.normal(15, 5, 1000)
axes[0, 2].hist(x2, bins=30)
axes[0, 2].set_xlim(0, 30)

np.random.seed(3)
x3 = 4 + np.random.normal(0, 2, 60)
y3 = 4 + np.random.normal(0, 2, len(x3))
axes[1, 0].scatter(x3, y3)
axes[1, 0].grid(True)

axes[1, 1].grid(True)

axes[1, 2].grid(True)

plt.show()
```
